```python
data = pd.read_csv(r'/kaggle/input/bikroycom-mobile-ad-price-dataset/mobile_data_v2.csv')
data
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(data['Name'])
plt.xticks(rotation=90)
plt.show()
```

> **It is clear that people here use Samsung, Vivo, Xiaomi Oppo more**

```python
plt.hist(data['Location'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.hist(data['Condition'])
plt.xticks(rotation=90)
plt.show()
```

```python
import seaborn as sns

pivot_table = data.pivot_table(index='Location', columns='Name', values='Condition', aggfunc='count', fill_value=0)

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='Blues', annot=True, fmt='d')
plt.xlabel('Mobile Brands')
plt.ylabel('Location')
plt.title('Mobile Brands Used in Each Location')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
average_price_by_location = data.groupby('Location')['Price'].mean().reset_index()

plt.figure(figsize=(12, 6))
plt.bar(average_price_by_location['Location'], average_price_by_location['Price'], color='skyblue')
plt.xlabel('Location')
plt.ylabel('Average Price')
plt.title('Average Prices of Mobile Brands in Each Location')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
average_price_by_brand = data.groupby('Name')['Price'].mean().reset_index()

plt.figure(figsize=(12, 6))
plt.bar(average_price_by_brand['Name'], average_price_by_brand['Price'], color='skyblue')
plt.xlabel('Mobile Brand')
plt.ylabel('Average Price')
plt.title('Average Prices of Mobile Brands')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
